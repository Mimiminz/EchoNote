```python
import whisper

model = whisper.load_model("base.en")
result = model.transcribe("romeo_act_4_shakespeare_64kb.mp3")
print(result["text"])
# print(result["segments"])
```

```python

f = open("romeo_output.txt", 'w')


for seg in result["segments"]:
  f.write(f"{seg['start']} {seg['end']}\n")
  f.write(seg["text"])

f.close()
```
